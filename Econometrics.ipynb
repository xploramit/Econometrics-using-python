# Question:

2. Compute the mean and standard deviation of **Age**.  
3. Standardize the **Age** variable.

P(\text{Age} > 50)

# Conceptual Flow

Step 1 ‚Üí Check if Age is roughly normal

Step 2 ‚Üí Compute mean and SD

Step 3 ‚Üí Standardize

Step 4 ‚Üí Convert 50 into Z-score

Step 5 ‚Üí Use normal CDF to compute probability


## Theoretical Framework

We calculate the Z-score:

$$
Z = \frac{50 - \mu}{\sigma}
$$

Then the required probability is:

$$
P(\text{Age} > 50) = 1 - \Phi(Z)
$$


$$
P(\text{Age} > 50) = 1 - \Phi\left(\frac{50 - \mu}{\sigma}\right)
$$

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Select Age and drop missing values
age = titanic["age"].dropna()

# Compute mean and standard deviation
mu = age.mean()
sigma = age.std()

# Create range of x values
x = np.linspace(age.min(), age.max(), 500)

# Compute Normal PDF
pdf = norm.pdf(x, mu, sigma)

plt.figure(figsize=(8,5))

# Plot histogram of actual age data
plt.hist(age, bins=30, density=True, alpha=0.4)

# Plot normal curve
plt.plot(x, pdf)
plt.axvline(50, linestyle='--')

plt.title("Normal PDF of Titanic Age")
plt.xlabel("Age")
plt.ylabel("Density")

plt.show()
```

```python
# Compute CDF
cdf = norm.cdf(x, mu, sigma)

plt.figure(figsize=(8,5))

plt.plot(x, cdf)
plt.axvline(50, linestyle='--')

plt.title("Normal CDF of Age")
plt.xlabel("Age")
plt.ylabel("Cumulative Probability")

plt.show()
```

```python
# ============================================
# Conceptual Flow of the Problem
# ============================================

# Step 1: Check if Age is roughly normally distributed.
#         (We can use histogram, KDE plot, or Q-Q plot.)

# Step 2: Compute the mean (mu) and standard deviation (sigma)
#         of the Age variable.

# Step 3: Standardize the Age variable using:
#         Z = (X - mu) / sigma

# Step 4: Convert the specific value 50 into a Z-score using:
#         Z_50 = (50 - mu) / sigma

# Step 5: Use the Normal Cumulative Distribution Function (CDF)
#         to calculate the probability:
#         P(Age > 50) = 1 - Œ¶(Z_50)

```

```python
# ============================================
# Titanic Dataset: Probability that Age > 50
# Using Normal Distribution + Z-score
# ============================================

# 1) Import required libraries
import pandas as pd              # for data handling
import numpy as np               # for numerical calculations
import seaborn as sns            # to load Titanic dataset easily
from scipy.stats import norm     # for Normal distribution CDF
```

```python
# ------------------------------------------------------------
# 2) Load the Titanic dataset
# ------------------------------------------------------------
titanic = sns.load_dataset("titanic")
```

```python
# (Optional) See the first few rows to understand the dataset
print("First 5 rows of Titanic dataset:")
print(titanic.head())
```

```python
# 3) Select the Age column and remove missing values (NaN)
#    Because mean/SD and probability calculations need real numbers
# ------------------------------------------------------------
age = titanic["age"].dropna()

print("\nHow many passengers have Age data (non-missing)?", len(age))
print("How many are missing Age?", titanic["age"].isna().sum())

```

```python
# ------------------------------------------------------------
# 4) Compute mean (mu) and standard deviation (sigma) of Age
# ------------------------------------------------------------
mu = age.mean()      # average age
sigma = age.std()    # standard deviation of age (spread)

print("\nMean Age (mu):", mu)
print("Standard Deviation (sigma):", sigma)
```

```python
# ------------------------------------------------------------
# 5) Standardize the Age values (convert to Z-scores)
#    Z = (X - mu) / sigma
#    This tells how many SDs each age is from the mean
# ------------------------------------------------------------
z_scores = (age - mu) / sigma

print("\nFirst 5 Z-scores:")
print(z_scores.head())
```

```python
# ------------------------------------------------------------
# 6) Compute the Z-score for Age = 50
#    This converts the value 50 into standard normal units
# ------------------------------------------------------------
z_50 = (50 - mu) / sigma

print("\nZ-score for Age = 50:", z_50)
```

```python
# ------------------------------------------------------------
# 7) Find probability that Age > 50 using the normal distribution
#    P(Age > 50) = 1 - P(Age <= 50)
#    For standard normal: P(Z > z_50) = 1 - Œ¶(z_50)
# ------------------------------------------------------------

prob = (1 - norm.cdf(z_50)) * 100

print(f"\nProbability that a passenger's Age > 50 (assuming normality): {prob:.2f}%")
```

# <center> OR </center>

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import seaborn as sns

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Drop missing age values
age = titanic['age'].dropna()

# Step 1: Compute mean and standard deviation
mu = age.mean()
sigma = age.std()

print("Mean Age:", mu)
print("Standard Deviation:", sigma)

# Step 2: Standardize Age
z_scores = (age - mu) / sigma

# Step 3: Calculate Z for age = 50
z_50 = (50 - mu) / sigma
print("Z-score for age 50:", z_50)

# Step 4: Probability P(Age > 50)
prob = 1 - norm.cdf(z_50)

print("Probability that Age > 50:", prob)
```

## Now,compare Theoretical and Empirical Probability

### Theoretical Probability

$$
P(\text{Age} > 50) = 1 - \Phi\left(\frac{50 - \mu}{\sigma}\right)
$$

---

### Empirical Probability

$$
\text{Empirical } P =
\frac{\text{Number of passengers with Age > 50}}
{\text{Total number of passengers with Age data}}
$$

```python
# Empirical Probability Calculation

empirical_prob = (age > 50).sum() / len(age)*100

print(f"Empirical Probability (Age > 50): {empirical_prob:.4f}%")
```

# <center> SKEWENESS </center>

A **normal distribution** is a **bell-shaped**, **symmetrical distribution** with a specific mathematical formula that describes how the data is spread out. **Skewness** indicates that the data is **not symmetrical**, which means it is not normally distributed.

**Skewness** is a measure of the **asymmetry** of a probability distribution. It is a statistical measure that describes the degree to which a dataset deviates from the **normal distribution**.

In a **symmetrical distribution**, the **mean**, **median**, and **mode** are all equal. In contrast, in a **skewed distribution**, the mean, median, and mode are not equal, and the distribution tends to have a **longer tail** on one side than the other.

Skewness can be **positive**, **negative**, or **zero**:

- **Positive skewness** means that the **tail is longer on the right side**.
- **Negative skewness** means that the **tail is longer on the left side**.
- **Zero skewness** indicates a **perfectly symmetrical distribution**.

## What is the Range of Skewness?

Skewness theoretically ranges from:

$$
-\infty \text{ to } +\infty
$$

But in practice, we interpret values as follows:

| Skewness Value | Interpretation |
|---------------|---------------|
| 0 | Perfectly symmetric |
| Between -0.5 and +0.5 | Approximately symmetric |
| Between 0.5 and 1 | Moderately skewed |
| Greater than 1 | Highly skewed |
| Between -1 and -0.5 | Moderately negatively skewed |
| Less than -1 | Highly negatively skewed |

---

## When is it Positive Skew?

Skewness > 0  

- Right tail is longer.  
- Mean > Median.  

**Example:**  
Income distribution ‚Äî a few very rich people pull the average upward.

---

## When is it Negative Skew?

Skewness < 0  

- Left tail is longer.  
- Mean < Median.  

**Example:**  
Exam where most students score high but a few fail badly.

---

## What Does 0.389 Tell Us?

Since:

$$
0 < 0.389 < 0.5
$$

 Titanic Age is approximately symmetric with slight right skew.

So assuming normality is not unreasonable.

Skewness measures asymmetry.

Positive value ‚Üí right tail longer.

Negative value ‚Üí left tail longer.

Zero ‚Üí symmetric.

# How to detect Outliers?

## Interpretation of  Œº + 3œÉ

This calculates:

$$
\mu + 3\sigma
$$

Where:

- $\mu$ = mean of Age  
- $\sigma$ = standard deviation of Age  

So this gives the value that is **3 standard deviations above the mean**.

---

## Why Is This Important?

In a normal distribution:

- 68% of the data lies within ¬±1œÉ  
- 95% lies within ¬±2œÉ  
- 99.7% lies within ¬±3œÉ  

This is called the **Empirical Rule (68‚Äì95‚Äì99.7 rule)**.

Therefore,

$$
\mu + 3\sigma
$$

is approximately the **upper extreme boundary** of the distribution.

```python
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generate a set of random data
x = np.random.uniform(low=0, high=1, size=1000)
```

```python
plt.hist(x)
```

```python
# Fit a uniform distribution to the data
params = stats.uniform.fit(x)
dist = stats.uniform(loc=params[0], scale=params[1])
```

```python
# Create a QQ plot of the data using the uniform distribution
fig = sm.qqplot(x, dist=dist, line='45')

# Add a title and labels to the plot
plt.title('QQ Plot of Uniform Distribution with Uniform Fit')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```

# Log Log Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the Pareto distribution
alpha = 3
xm = 1

# Create an array of x values
x = np.linspace(0.1, 10, 1000)

# Calculate the y values of the Pareto distribution
y = alpha * (xm**alpha) / (x**(alpha+1))


plt.plot(x,y)
```

```python
# Create the log-log plot
plt.plot(np.log(x),np.log(y))

# Add labels and a title
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Pareto distribution')

# Show the plot
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Define the parameters of the Pareto distribution
alpha = 2
xm = 1

# Generate a set of random data from the Pareto distribution
x = stats.pareto.rvs(b=alpha, scale=xm, size=1000)
```

```python
plt.hist(x)
```

```python
# Fit a Pareto distribution to the data
params = stats.pareto.fit(x, floc=0)

# Create a Pareto distribution object with the fitted parameters
dist = stats.pareto(b=params[0], scale=params[2])
```

```python
# Create a QQ plot of the data using the Pareto distribution
fig = sm.qqplot(x, dist=dist, line='45')

# Add a title and labels to the plot
plt.title('QQ Plot of Pareto Distribution with Pareto Fit')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```
