# <center> PROBABILITY DISTRIBNUTION FUNCTIONS  </center> 

```python
import pandas as pd
```

```python
import random
```

```python
L = []
for i in range(10000):
    L.append(random.randint(1,6))
```

```python
# Checking the length of L using len()
```

```python
len(L)
```

```python
# Checking 1st 10 randomly generated nos.
```

```python
L[:10]
```

```python
# Converting L in a series using Series function
```

```python
pd.Series(L)
```

```python
# let's check 1,2,3,4,5, 6 kitni baar aaya? (using value_counts() function)
```

```python
pd.Series(L).value_counts()
```

```python
# ab in sabkaa probablity nikaalte hai by dividing the total count by total value we have i.e., 10000 and store in s
```

```python
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum())
```

```python
s
```

# ab nos. ko sort kr dete hai by using <b>sort_index()</b>

```python
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
s
```

```python
# ab dekho roughly sabkaa ie., 1,2,3,4,5,6 sabkaa probablities almost same hai
```

# ab isko plot karke dekhte hai to get the <b>Probablity Mass Function </b> using plot and kind = bar

```python
s.plot(kind='bar')
```

```python
# upar bar diagram ko uniform discrete random distribution bhi kehte hai
```

# Ab 2 dice run krte hai and then calculate its sum as well

```python
import pandas as pd
```

```python
import random
```

```python
L = []
for i in range(10000):
  a = random.randint(1,6)
  b = random.randint(1,6)

  L.append(a + b)
```

```python
len(L)
```

```python
# ab sum dekhte hai 1st 5 ka
```

```python
L[:5]
```

```python
# ab probabilities dekhte hai sum ka
```

```python
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
s
```

```python
# ab cumilative sum dekh sakte hai probabilities ka using cumsum()
```

```python
import numpy as np
np.cumsum(s)
```

# ab is sorted probabilities ka graph plot karte hai (PMY)

```python
s.plot(kind='bar')
```

```python
# This distribution looks like a normal distribution function
```

# CUMULATIVE DISTRIBUTION FUNCTION(CDF) OF PMF

```python
np.cumsum(s).plot(kind='bar')
```
