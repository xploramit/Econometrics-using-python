# <center> NORMAL DISTRIBUTION </center>

## What is a Normal Distribution?

A **normal distribution** is a **bell-shaped, symmetrical probability distribution** in which most of the values are concentrated around the **mean (average)**, and fewer values appear as we move away from the center.

In a normal distribution:

- The **mean**, **median**, and **mode** are equal.
- The distribution is **perfectly symmetric** around the mean.
- The total area under the curve equals **1** (which represents total probability).

---

## Example

Suppose we measure the **heights of students** in a college.

- Most students will have heights close to the average height.
- Very few students will be extremely short or extremely tall.

If we plot these heights on a graph, we get a **bell-shaped curve**, which represents a normal distribution.

Another common example is:
- **Marks in a large competitive exam**
- **Measurement errors**
- **IQ scores**

---

## Mathematical Formula of Normal Distribution

The probability density function (PDF) of a normal distribution is:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)
$$

Where:

- $\mu$ = Mean of the distribution  
- $\sigma$ = Standard deviation (measure of spread)  
- $\sigma^2$ = Variance  
- $\pi \approx 3.1416$  
- $e \approx 2.718$  

---

## Important Properties

- 68% of data lies within $\mu \pm 1\sigma$
- 95% lies within $\mu \pm 2\sigma$
- 99.7% lies within $\mu \pm 3\sigma$

This is known as the **Empirical Rule (68‚Äì95‚Äì99.7 rule)**.

---

In simple words:

A **normal distribution** shows how data is spread out when most values are near the average and extreme values are rare.

# Standardize Normal Distribution

## Why Do We Standardize a Normal Distribution?

Standardizing a normal distribution means converting the original variable into a **standard normal variable (Z-score)**.

### Why Do We Standardize?

We standardize for three main reasons:

### 1Ô∏è‚É£ To Make Different Data Comparable
Different datasets may have different units:
- Height (in cm)
- Income (in rupees)
- Marks (out of 100)

By standardizing, we remove the units and convert everything into a common scale.

---

### 2Ô∏è‚É£ To Use Standard Normal Tables

Instead of calculating probability separately for every mean and standard deviation, we convert everything into a **standard normal distribution** where:

- Mean = 0
- Standard deviation = 1

Then we can use:
- Z-tables
- Statistical software
- Standard normal CDF

This makes probability calculation easy.

---

### 3Ô∏è‚É£ To See How Far a Value Is from the Mean

Standardization tells us:

> How many standard deviations a value is away from the mean.

This helps detect:
- Extreme values
- Outliers
- Unusual observations

---

## How Do We Standardize?

We use the formula:

$$
Z = \frac{X - \mu}{\sigma}
$$

Where:

- $X$ = Original value  
- $\mu$ = Mean  
- $\sigma$ = Standard deviation  

---

## What Does This Formula Do?

1. Subtract the mean ‚Üí centers the data at 0  
2. Divide by standard deviation ‚Üí scales the spread to 1  

After this transformation:

- New mean = 0  
- New standard deviation = 1  

This new variable is called the **Standard Normal Distribution (Z-distribution)**.

---

## In Simple Words

Standardization:

- Removes units  
- Makes comparison easier  
- Simplifies probability calculation  
- Helps detect unusual values  

It converts any normal distribution into a **common standard scale**.

## Problem (Normal Distribution + Z-table)

Suppose the heights of adult males in a certain population follow a **normal distribution** with:

- Mean: $\mu = 68$ inches  
- Standard deviation: $\sigma = 3$ inches  

**Question:** What is the probability that a randomly selected adult male is **taller than 72 inches**?

We want to find:

$$
P(X > 72)
$$

---

## Step 1: Convert 72 into a Z-score (Standardize)

Use the Z-score formula:

$$
Z = \frac{X - \mu}{\sigma}
$$

Substitute values:

$$
Z = \frac{72 - 68}{3} = \frac{4}{3} = 1.333\approx 1.33
$$

So,

$$
P(X > 72) = P(Z > 1.33)
$$

---

## Step 2: Use the Z-table to find $P(Z \le 1.33)$

Most Z-tables give the **left-side (cumulative) probability**:

$$
P(Z \le z)
$$

Now locate **1.33** in the Z-table:

- Row = **1.3**
- Column = **0.03**

From the Z-table:

$$
P(Z \le 1.33) \approx 0.9082
$$

(This means 90.82% of values lie **below** 1.33.)

---

## Step 3: Convert to the required probability $P(Z > 1.33)$

We need the **right tail**:

$$
P(Z > 1.33) = 1 - P(Z \le 1.33)
$$

So:

$$
P(Z > 1.33) = 1 - 0.9082 = 0.0918
$$

---

## Final Answer

$$
P(X > 72) \approx 0.0918
$$

 **Interpretation:**  
There is about a **9.18%** chance that a randomly selected adult male is taller than **72 inches**.

---

## Quick Summary

1. Standardize: $Z = (72-68)/3 = 1.33$  
2. Z-table: $P(Z \le 1.33)=0.9082$  
3. Right tail: $P(Z > 1.33)=1-0.9082=0.0918$

# That‚Äôs a very good question by Sanjana.

Yes, the **Z-table** is essentially a **Cumulative Distribution Function (CDF) table**, but specifically for the **standard normal distribution**.

### What is a CDF?

A **Cumulative Distribution Function (CDF)** gives:

$$
P(X \le x)
$$

This means the probability that a random variable is **less than or equal to** a particular value.

---

### What Does the Z-table Give?

The Z-table provides:

$$
P(Z \le z)
$$

Where:

- $Z \sim N(0,1)$
- Mean = 0  
- Standard deviation = 1  

It gives the **cumulative probability from negative infinity up to that Z value**.

---

### Example

If we look up:

$$
Z = 1.33
$$

The Z-table gives:

$$
P(Z \le 1.33) = 0.90824
$$

# Solving it by manually

## Titanic Dataset: Calculating $P(\text{Age} > 50)$ by Hand (Using Z-score + Z-table)

We assume **Age** is approximately normally distributed:

$$
X \sim N(\mu, \sigma^2)
$$

Our goal is to calculate:

$$
P(X > 50)
$$

---

## Step 1: Compute Mean and Standard Deviation (from Titanic Age data)

From the Titanic dataset (after removing missing ages), suppose we get:

- Mean age: $\mu$
- Standard deviation: $\sigma$

*(In your notebook, you will compute these using Python, but for hand calculation we treat them as known values.)*

---

## Step 2: Standardize (Convert 50 into a Z-score)

Use the standardization formula:

$$
Z = \frac{X - \mu}{\sigma}
$$

For \(X = 50\):

$$
Z_{50} = \frac{50 - \mu}{\sigma}
$$

This tells us how many **standard deviations** 50 is away from the mean.

---

## Step 3: Convert the Probability into Standard Normal Form

Because \(Z\) follows a standard normal distribution:

$$
P(X > 50) = P\left(Z > \frac{50 - \mu}{\sigma}\right)
$$

So:

$$
P(X > 50) = P(Z > Z_{50})
$$

---

## Step 4: Use Z-table (CDF table)

Most Z-tables give:

$$
P(Z \le z)
$$

So first find:

$$
P(Z \le Z_{50})
$$

from the Z-table.

---

## Step 5: Convert Left-tail Probability to Right-tail Probability

We need \(P(Z > Z_{50})\), which is the **right tail**, so:

$$
P(Z > Z_{50}) = 1 - P(Z \le Z_{50})
$$

---

## Hand Calculation of $P(\text{Age} > 50)$

---

From the Z-table:

This represents the **left-tail probability**.

P(\text{Age} > 50)

Which means:

P(Z > 1.398)

P(Z > 1.398) = 1 - 0.9177

```python
# 3) Select the Age column and remove missing values (NaN)
#    Because mean/SD and probability calculations need real numbers
# ------------------------------------------------------------
age = titanic["age"].dropna()

print("\nHow many passengers have Age data (non-missing)?", len(age))
print("How many are missing Age?", titanic["age"].isna().sum())

```

```python
# ------------------------------------------------------------
# 4) Compute mean (mu) and standard deviation (sigma) of Age
# ------------------------------------------------------------
mu = age.mean()      # average age
sigma = age.std()    # standard deviation of age (spread)

print("\nMean Age (mu):", mu)
print("Standard Deviation (sigma):", sigma)
```

```python
# ------------------------------------------------------------
# 5) Standardize the Age values (convert to Z-scores)
#    Z = (X - mu) / sigma
#    This tells how many SDs each age is from the mean
# ------------------------------------------------------------
z_scores = (age - mu) / sigma

print("\nFirst 5 Z-scores:")
print(z_scores.head())
```

```python
# ------------------------------------------------------------
# 6) Compute the Z-score for Age = 50
#    This converts the value 50 into standard normal units
# ------------------------------------------------------------
z_50 = (50 - mu) / sigma

print("\nZ-score for Age = 50:", z_50)
```

# <center> OR </center>

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import seaborn as sns

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Drop missing age values
age = titanic['age'].dropna()

# Step 1: Compute mean and standard deviation
mu = age.mean()
sigma = age.std()

print("Mean Age:", mu)
print("Standard Deviation:", sigma)

# Step 2: Standardize Age
z_scores = (age - mu) / sigma

# Step 3: Calculate Z for age = 50
z_50 = (50 - mu) / sigma
print("Z-score for age 50:", z_50)

# Step 4: Probability P(Age > 50)
prob = 1 - norm.cdf(z_50)

print("Probability that Age > 50:", prob)
```

# <center> SKEWENESS </center>

A **normal distribution** is a **bell-shaped**, **symmetrical distribution** with a specific mathematical formula that describes how the data is spread out. **Skewness** indicates that the data is **not symmetrical**, which means it is not normally distributed.

**Skewness** is a measure of the **asymmetry** of a probability distribution. It is a statistical measure that describes the degree to which a dataset deviates from the **normal distribution**.

In a **symmetrical distribution**, the **mean**, **median**, and **mode** are all equal. In contrast, in a **skewed distribution**, the mean, median, and mode are not equal, and the distribution tends to have a **longer tail** on one side than the other.

Skewness can be **positive**, **negative**, or **zero**:

- **Positive skewness** means that the **tail is longer on the right side**.
- **Negative skewness** means that the **tail is longer on the left side**.
- **Zero skewness** indicates a **perfectly symmetrical distribution**.

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
titanic['Age'].skew()
```

# The skewness value is:  0.389

What does this mean?

The Age distribution is slightly positively skewed.

There are a few older passengers pulling the distribution slightly to the right.

But since 0.389 is small, the distribution is **close to symmetric**.


## What is the Range of Skewness?

Skewness theoretically ranges from:

$$
-\infty \text{ to } +\infty
$$

But in practice, we interpret values as follows:

| Skewness Value | Interpretation |
|---------------|---------------|
| 0 | Perfectly symmetric |
| Between -0.5 and +0.5 | Approximately symmetric |
| Between 0.5 and 1 | Moderately skewed |
| Greater than 1 | Highly skewed |
| Between -1 and -0.5 | Moderately negatively skewed |
| Less than -1 | Highly negatively skewed |

---

## When is it Positive Skew?

Skewness > 0  

- Right tail is longer.  
- Mean > Median.  

**Example:**  
Income distribution ‚Äî a few very rich people pull the average upward.

---

## When is it Negative Skew?

Skewness < 0  

- Left tail is longer.  
- Mean < Median.  

**Example:**  
Exam where most students score high but a few fail badly.

---

## What Does 0.389 Tell Us?

Since:

$$
0 < 0.389 < 0.5
$$

 Titanic Age is approximately symmetric with slight right skew.

So assuming normality is not unreasonable.

Skewness measures asymmetry.

Positive value ‚Üí right tail longer.

Negative value ‚Üí left tail longer.

Zero ‚Üí symmetric.
