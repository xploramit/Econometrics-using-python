# üé≤ Simulation of Two Dice Rolls (10,000 Times)



Step 1: Create an Empty List

We first create an **empty list**:

**L = [ ]**

This list L will store the sums of two dice rolls.

Right now, the list is empty: [ ]

Step 2: Start a **Loop** That Repeats 10,000 Times

We use:

**for** i in range(10000):

The command range(10000) means:

0, 1, 2, 3, ..., 9999

So the loop runs 10,000 times.

The variable i is just a counter.
We are not using it inside the loop ‚Äî it only helps repeat the process 10,000 times.

Step 3: Roll the First Die

Inside the loop, we write:

a = random.randint(1,6)

The function **randint(1,6)** generates a random integer from:

1, 2, 3, 4, 5, 6

So the variable a behaves like the outcome of the first die.

Example:

a = 4

Step 4: Roll the Second Die

Next, we write:

b = random.randint(1,6)

This again generates a random integer from 1 to 6.

This time it represents the second die.

Example:

b = 2

Step 5: Add the Two Dice and Store the Sum

We then write:

**L.append(a + b)**

Here:

a + b gives the sum of the two dice.

Possible sums range from:

2 (1+1) to 12 (6+6)

The function **append()** means:

‚ÄúAdd the value to the end of the list.‚Äù

Example:

If
a = 4
b = 2

Then:

a + b = 6

So we store 6 in the list L.

After the Loop Finishes

After running the loop 10,000 times:

The list L will contain 10,000 numbers.

Each number represents the sum of two dice rolls.

Each value will lie between 2 and 12.

Example (first few values of L):

[7, 9, 5, 12, 6, 8, 7, 4, ...]

```python
L = []
for i in range(10000):
  a = random.randint(1,6)
  b = random.randint(1,6)

  L.append(a + b)
```

# KDE

```python
# generate a sample
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = np.hstack((sample1, sample2))
```

```python
sample
```

```python
# plot histogram bins=50
plt.hist(sample,bins=50)
```

```python
from sklearn.neighbors import KernelDensity

model = KernelDensity(bandwidth=5, kernel='gaussian')

# convert data to a 2D array
sample = sample.reshape((len(sample), 1))

model.fit(sample)
```

```python
values = np.linspace(sample.min(),sample.max(),100)
values = values.reshape((len(values), 1))
```

```python
probabilities = model.score_samples(values)
probabilities = np.exp(probabilities)
```

```python
# score_samples(values) returns the log-density estimate of the input samples values. This is because the score_samples() method of the 
# KernelDensity class returns the logarithm of the probability density estimate rather than the actual probability density estimate.
```

```python
plt.hist(sample, bins=50, density=True)
plt.plot(values[:], probabilities)
plt.show()
```

```python
sns.kdeplot(sample.reshape(1000),bw_adjust=0.02)
```

We plotted a **countplot**:

Blue (0) ‚Üí Did NOT survive

Now let us interpret it clearly and properly.

**Interpretation of the Graph**
1) Third Class (Pclass = 3)

Very high number of non-survivors (blue bar is extremely tall)

Much lower number of survivors (orange bar smaller)

This tells us most third-class passengers did not survive. This class suffered the highest mortality.

2) First Class (Pclass = 1)

Survivors (orange) are higher than non-survivors (blue)

Survival appears relatively strong. This means that First-class passengers had a higher chance of survival.

3) Second Class (Pclass = 2)

Survivors and non-survivors are closer. Slightly more non-survivors. This class lies between 1st and 3rd class in survival probability.

**Main Conclusion**

Survival probability clearly depends on passenger class. It follows this pattern:

**P(Survival‚à£1st) > P(Survival‚à£2nd) > P(Survival‚à£3rd)**

**So survival was not random. It was strongly associated with socioeconomic status.**

**Economic Interpretation**

This reflects:

Social hierarchy

Unequal access to lifeboats

Class-based structural advantage

Economic inequality during crisis

But, this graph shows counts, not probabilities.

Since there are more third-class passengers overall, the blue bar is naturally larger.

For proper econometric interpretation, you should look at proportions:

pd.crosstab(titanic['Pclass'], titanic['Survived'], normalize='index')

That gives conditional probabilities:

ùëÉ(Survive‚à£Class)

Which is statistically more correct.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

ax = sns.countplot(data=titanic, x='Pclass', hue='Survived')

# Add counts on top of bars
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}',
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom')

plt.show()
```


 What This Does (Simple Explanation)

ax.patches ‚Üí goes through each bar

p.get_height() ‚Üí gets the count value

annotate() ‚Üí writes the number above the bar

ha='center' ‚Üí centers text

va='bottom' ‚Üí places it above the bar

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'], normalize='index')
```

 Step 0: What is the goal?

We want a table that tells us:

For each passenger class (1, 2, 3), what fraction survived and what fraction did not survive.

In probability language:

P(Survived ‚à£ Pclass)
Step 1: pd.crosstab(...)

pd.crosstab() means cross-tabulation (like a contingency table).

It creates a table of counts (or proportions) between two variables.

Step 2: First input = rows

titanic['Pclass']

This becomes the row categories.

So rows will be:

Class 1

Class 2

Class 3

Step 3: Second input = columns

titanic['Survived']

This becomes the column categories.

Usually:

0 = Did not survive

1 = Survived

So columns will be:

0

1

Step 4: What happens without normalization?

pd.crosstab(titanic['Pclass'], titanic['Survived'])

You would get counts like:

Pclass	  0	        1
1	   (count)	 (count)
2	   (count)	 (count)
3	   (count)	 (count)
Step 5: normalize='index' (most important part)

normalize='index'

This converts counts into row-wise proportions.

Meaning: Each row (each class) will sum to 1. It shows percentages within that class

So for each Pclass:

Proportion Not Survived + Proportion Survived = 1

Example interpretation:

If row for Class 1 shows 0: 0.37 and 1: 0.63

It means: 37% died, 63% survived in 1st class.

**Final Interpretation** (what this table tells us)

This table answers:

‚ÄúAmong 1st class passengers, what share survived?‚Äù

‚ÄúAmong 2nd class passengers, what share survived?‚Äù

‚ÄúAmong 3rd class passengers, what share survived?‚Äù

This is much better than just counts, because it compares survival fairly across classes.

# How to show as percentages?

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'], normalize='index') * 100
```
