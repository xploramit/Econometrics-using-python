# Ab 2 dice run krte hai and then calculate its sum as well

```python
import pandas as pd
```

```python
import random
```

```python
L = []
for i in range(10000):
  a = random.randint(1,6)
  b = random.randint(1,6)

  L.append(a + b)
```

```python
len(L)
```

```python
# ab sum dekhte hai 1st 5 ka
```

```python
L[:5]
```

```python
# ab probabilities dekhte hai sum ka
```

```python
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
s
```

```python
# ab cumilative sum dekh sakte hai probabilities ka using cumsum()
```

```python
import numpy as np
np.cumsum(s)
```

# ab is sorted probabilities ka graph plot karte hai (PMY)

```python
s.plot(kind='bar')
```

```python
# This distribution looks like a normal distribution function
```

# CUMULATIVE DISTRIBUTION FUNCTION(CDF) OF PMF

```python
np.cumsum(s).plot(kind='bar')
```

# KDE

```python
# generate a sample
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = np.hstack((sample1, sample2))
```

```python
sample
```

```python
# plot histogram bins=50
plt.hist(sample,bins=50)
```

```python
from sklearn.neighbors import KernelDensity

model = KernelDensity(bandwidth=5, kernel='gaussian')

# convert data to a 2D array
sample = sample.reshape((len(sample), 1))

model.fit(sample)
```

```python
values = np.linspace(sample.min(),sample.max(),100)
values = values.reshape((len(values), 1))
```

```python
probabilities = model.score_samples(values)
probabilities = np.exp(probabilities)
```

```python
# score_samples(values) returns the log-density estimate of the input samples values. This is because the score_samples() method of the 
# KernelDensity class returns the logarithm of the probability density estimate rather than the actual probability density estimate.
```

```python
plt.hist(sample, bins=50, density=True)
plt.plot(values[:], probabilities)
plt.show()
```

```python
sns.kdeplot(sample.reshape(1000),bw_adjust=0.02)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1) Clean + ensure correct dtypes
plot_df = df[['petal_width', 'species']].copy()
plot_df['petal_width'] = pd.to_numeric(plot_df['petal_width'], errors='coerce')
plot_df['species'] = plot_df['species'].astype('category')
plot_df = plot_df.dropna(subset=['petal_width', 'species'])

# 2) KDE plot with hue
plt.figure(figsize=(7,4))
sns.kdeplot(data=plot_df, x='petal_width', hue='species', fill=True, common_norm=False)
plt.title('KDE of Petal Width by Species')
plt.show()

# 3) ECDF plot with hue
plt.figure(figsize=(7,4))
sns.ecdfplot(data=plot_df, x='petal_width', hue='species')
plt.title('ECDF of Petal Width by Species')

plt.show()
```
